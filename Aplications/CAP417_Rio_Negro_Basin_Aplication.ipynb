# 3.Hypothesis Test

```python
from math import sqrt
from scipy.stats import norm

def calc_mean(array):
  return np.mean(array)

def calc_std(array):
  return np.std(array)

#Function that returns a subset of size s of the population p
def calc_subset(p, s, seed):
  np.random.seed(seed)
  return np.random.choice(p, s, replace=False)  
```

```python
#Sample Size 
n_s = 100 

seed=1
subset = calc_subset(manuel_Diaz_precipitation, n_s, seed)

#Sample Average
mu_s = calc_mean(subset) 
#Sample Standard Deviation
sigma_s = calc_std(subset)
#Confidence coefficient
cc = 0.95

#Calculating the significance level
alpha = 1 - cc

#Normal parameters
q = 1 - alpha/2
mu_z = 0
sigma_z = 1
#ppf(q, loc=0, scale=1)
z_value = norm.ppf(q, loc=mu_z, scale=sigma_z) 

#Margin of error
margin = z_value * sigma_s / sqrt(n_s)

#Confidence interval for the mean
mu_p_inf = mu_s - margin
mu_p_sup = mu_s + margin

print("Subset", subset)
print("Sample Mean: ", mu_s)
print("Population Mean: ", calc_mean(manuel_Diaz_precipitation))


print("\n\nSolution to this problem: ")
print("\nZ = ", z_value)
print("\n95% confidence interval for the mean:")
print("\n\tLower: ", mu_p_inf)
print("\n\tUpper: ", mu_p_sup)
```

## 4.2 Systematic sampling

Systematic sampling is a type of probability sampling method in which sample members from a larger population are selected according to a random starting point but with a fixed, periodic interval. This interval, called the sampling interval, is calculated by dividing the population size by the desired sample size.

```python
import random
```

```python
def systematic_sampling(dataset, samples):
  interval = len(dataset) // samples
  random.seed(1)
  start = random.randint(0, interval)
  index = np.arange(start, len(dataset), step = interval)
  systematic_sampling = dataset.iloc[index]
  return systematic_sampling
```

```python
df_systematic_sampling_level = systematic_sampling(df_level, 100)
df_systematic_sampling_rain = systematic_sampling(df_rain, 100)
df_systematic_sampling_level. shape, df_systematic_sampling_rain.shape
```

## 4.3 Cluster Sampling

In cluster sampling, researchers divide a population into smaller groups known as clusters. They then randomly select among these clusters to form a sample. Cluster sampling is a method of probability sampling that is often used to study large populations, particularly those that are widely geographically dispersed

```python
def cluster_sampling(dataset, cluster_number):
  interval = len(dataset) / cluster_number

  cluster = []
  id_cluster = 0
  count = 0
  for _ in dataset.iterrows():
    cluster.append(id_cluster)
    count += 1
    if count > interval:
      count = 0
      id_cluster += 1

  dataset['cluster'] = cluster
  random.seed(1)
  selected_cluster = random.randint(0, cluster_number)
  return dataset[dataset['cluster'] == selected_cluster]
```

## 4.5 Reservoir sampling

Reservoir sampling is a family of randomized algorithms for choosing a simple random sample, without replacement, of k items from a population of unknown size n in a single pass over the items. The size of the population n is not known to the algorithm and is typically too large for all n items to fit into main memory. The population is revealed to the algorithm over time, and the algorithm cannot look back at previous items. At any point, the current state of the algorithm must permit extraction of a simple random sample without replacement of size k over the part of the population seen so far

```python
def reservoir_sampling(dataset, samples):
  stream = []
  for i in range(len(dataset)):
    stream.append(i)

  i = 0
  size = len(dataset)

  reservoir = [0] * samples
  for i in range(samples):
    reservoir[i] = stream[i]

  while i < size:
    j = random.randrange(i + 1)
    if j < samples:
      reservoir[j] = stream[i]
    i += 1

  return dataset.iloc[reservoir]
```

```python
def plotSdt(axs, title, sample):
    # set data
    bins = 30
    axs.hist(sample, bins, density=True, color ='green')
    
    # set text box and mean line
    mu = np.mean(sample)
    median = np.median(sample)
    sigma = np.std(sample)
    lenth = len(sample)
    textstr = '\n'.join((r'$\mu=%.2f$' % (mu, ), r'$\mathrm{median}=%.2f$' % (median, ), r'$\sigma=%.2f$' % (sigma, ), r'qdt=%.0f' % (lenth, )))
    props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
    axs.text(1, 0.5, textstr, transform=axs.transAxes, fontsize=14, verticalalignment='bottom', bbox=props)  
    axs.axvline(mu, color='k', linestyle='dashed', linewidth=1)

    #Set bottom and left spines as x and y axes of coordinate system
    axs.spines['bottom'].set_position('zero')
    axs.spines['left'].set_position('zero')
    
    #Remove top and right spines
    axs.spines['top'].set_visible(False)
    axs.spines['right'].set_visible(False)

    
    #Create labels 
    axs.set_xlabel('Level', size=14)
    axs.set_ylabel('Frequency', size=14)

    # set axes limits
    axs.set_xlim(0, 4.5)
    axs.set_ylim(0, 1.4)

    # set title
    axs.set_title(title, pad=20)
    return
```
